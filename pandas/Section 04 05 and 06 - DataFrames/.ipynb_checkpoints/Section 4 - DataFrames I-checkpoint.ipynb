```python
# Pull DF in reverse column order
nba[nba.columns[::-1]].head(1)
```

```python
# Add new column called Sport
# This OVERWRITES if column exists
nba["Sport"] = "Basketball"
nba["League"] = "NBA"
```

```python
# Go back to original nba DF
nba = nba[["Name", "Team", "Number", "Position", "Age", "Height", "Weight", "College", "Salary"]]
```

```python
# Insert method to create new column
# (loc, column, value)
nba.insert(3, column = "Sport", value = "Basketball")
nba.insert(1, column = "League", value = "National Basketball Association")
# nba.head(2)
```

```python
# Add to every row of a certain column
# If value is NaN, it will stay NaN
nba["Age"].add(5)

# Can also do add with math notation
nba["Age"] + 5

# Subtract
nba["Salary"].sub(100)
nba["Salary"] - 150

# Multiply
nba["Weight"].mul(.453592)

# Can add to DF as well
nba.insert(list(nba.columns.values).index("Weight") + 1, column = "Weight in Kg", value = nba["Weight"].mul(.453592))
nba.head(1)
```

```python
# Convert salary to millions
nba["Salary"].div(1000000)

nba.insert(list(nba.columns.values).index("Salary") + 1, column = "Salary in Millions ($)", value = nba["Salary"].div(1000000))
nba.head(5)
```

```python
# Get counts of players on each team
nba["Team"].value_counts()
```

```python
# Drop rows with a null value

# This drops all rows with any Nulls at all
# nba.dropna()

# This drops all rows with Null in all columns
# nba.dropna(how = "all", inplace = True)

# This drops all columns with Null in any rows
# nba.dropna(1)

nba.dropna(subset = ["Name", "Team", "Number", "Position", "Age", "Height", "Weight", "College", "Salary"], how = "all", inplace = True)
```

```python
# Fill NULL Values with the .fillna() Method

# Fills every column that is NaN with 0
# nba.fillna(0)

# Fill just the salary column NaN rows with 0
nba["Salary"].fillna(0, inplace = True)
nba["Salary in Millions ($)"].fillna(0, inplace = True)

# Set college to No College when College value is NaN
nba["College"].fillna("No College", inplace = True)
```

```python
# Change data types using the .astype() Method

# Use info to show if the columns are non-null
# nba.info()

# use .astype to change salary to an int (There is not inplace param in .astype)
nba["Salary"] = nba["Salary"].astype("int")

# Convert age and number to int as well
nba["Number"] = nba["Number"].astype("int")
nba["Age"] = nba["Age"].astype("int")

# Category datatypes can save memory by changing from objects to smaller list of unique categories
nba["Position"] = nba["Position"].astype("category")
nba["Team"] = nba["Team"].astype("category")
nba["League"] = nba["League"].astype("category")
```

```python
# Sort a DF with the .sort_values() Method

# Sort by name
# nba.sort_values("Name")

# Sort by age oldest to youngest
nba.sort_values("Age", ascending = False)

Output = False
```

```python
# Sort a DF by multiple columns

nba.sort_values(["Team", "Name"], ascending = [False, True])

Output = False
```

```python
# Sort a DF with the .sort_index() Method
nba.sort_index(ascending = False)

Output = False
```

```python
# Rank values of a DF with .rank() Method

nba["Salary Rank"] = nba["Salary"].rank(ascending = False).astype("int")
```

```python
# Show team salaries in descending order

nba[["Team", "Salary"]].groupby("Team").agg('sum').sort_values("Salary", ascending = False)
```
