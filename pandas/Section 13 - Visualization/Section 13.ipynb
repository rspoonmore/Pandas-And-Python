```python
import pandas as pd
from pandas_datareader import data
import datetime as dt
import matplotlib.pyplot as plt

# To view displays in notebook instead as popups
%matplotlib inline
```

```python
bb = data.DataReader("BB", data_source="yahoo", start="2000-01-01", end="2021-12-31")
```

```python
# To plot every column in DF
# bb.plot()
# To plot specific column in DF
# bb["Close"].plot()
# Can also set y
# bb.plot(y="Close")
bb[["High", "Close"]].plot()
# pd.Series(bb["High"] - bb["Close"]).plot()
```

```python
# Modifying plot aesthetics
# plt.style.available
# plt.style.use("fivethirtyeight")
plt.style.use("ggplot")
bb.plot(y="Close")
```

```python
# Creating bar graphs
def rank_perf(op, cl):
    if op < cl:
        return "Increase"
    elif op > cl:
        return "Decrease"
    elif op == cl:
        return "Stagnant"
    else:
        return "Unknown" 


# Simple Bar of counts    
simpleBar = bb.copy()
simpleBar.apply(lambda row: rank_perf(row["Open"], row["Close"]), axis=1).value_counts().plot(kind="bar")

# Horizontal bar by year since 18
hBar = bb.copy().reset_index()
hBar["Year"] = hBar["Date"].dt.year
hBar = hBar[hBar["Year"] >= 2018]
hBar.set_index("Year", inplace=True)
hBar["Performance"] = hBar.apply(lambda row: rank_perf(row["Open"], row["Close"]), axis=1)
hBarData = hBar.groupby(["Year", "Performance"])["Date"].count()
hBarData.unstack("Performance").sort_index(ascending=False).plot(kind="barh")
```

```python
# Pie charts
def compareToMean(price):
    mean = bb["Close"].mean()
    if price > mean:
        return "Above Average"
    elif price < mean:
        return "Below Average"
    elif price == mean:
        return "At Average"
    else:
        return "Unknown"

bbPie = bb["Close"].apply(compareToMean).value_counts()
# Simple Pie
# bbPie.plot(kind="pie", label="", title="Performance")

# Complex Pie
labels = list(bbPie.index)
explode = [0, .1]
bb["Close"].apply(compareToMean).value_counts().plot.pie(labels=labels,
                                                         explode=explode,
                                                         shadow=True,
                                                         autopct=lambda x: "{:.2f}%".format(x),
                                                         label="",
                                                         title="Performance"
                                                        )
```
